```python

class ArbreBinaire:
   def __init__(self, valeur):
      self.valeur = valeur
      self.enfant_gauche = None
      self.enfant_droit = None

   def insert_gauche(self, valeur):
      if self.enfant_gauche == None:
         self.enfant_gauche = ArbreBinaire(valeur)
      else:
         new_node = ArbreBinaire(valeur)
         new_node.enfant_gauche = self.enfant_gauche
         self.enfant_gauche = new_node

   def insert_droit(self, valeur):
      if self.enfant_droit == None:
         self.enfant_droit = ArbreBinaire(valeur)
      else:
         new_node = ArbreBinaire(valeur)
         new_node.enfant_droit = self.enfant_droit
         self.enfant_droit = new_node

   def get_valeur(self):
      return self.valeur

   def get_gauche(self):
      return self.enfant_gauche

   def get_droit(self):
      return self.enfant_droit

```

```python
racine = ArbreBinaire('Moi')
racine.insert_gauche('arthur')
racine.insert_droit('andrio')

b_node = racine.get_gauche()
b_node.insert_gauche('jean')
b_node.insert_droit('roger')

f_node = racine.get_droit()
f_node.insert_gauche('mauricette')
f_node.insert_droit('berth')

c_node = b_node.get_gauche()
c_node.insert_droit('sylvaine')

g_node = f_node.get_gauche()
g_node.insert_gauche('jacky')

h_node = f_node.get_droit()
h_node.insert_droit('charly')
```

```python
def affiche(node):
   if node != None:
      return (node.get_valeur(),affiche(node.get_gauche()),affiche(node.get_droit()))
```

```python
affiche(racine)
```

```python
def afficher_arbre(noeud, niveau=0):
    if noeud is None:
        return
    

    print("   " * niveau + str(noeud.get_valeur()))
    
    afficher_arbre(noeud.get_gauche(), niveau + 1)
    afficher_arbre(noeud.get_droit(), niveau + 1)
```

```python
afficher_arbre(racine)
```
